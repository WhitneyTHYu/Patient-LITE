```python
import pandas as pd

disease_lst = ['Alzheimer', 'Diabetes', 'NSCLC', 'UC', 'AML', 'CF', 'MS']
input_path = f'/Users/whitneyyu/Documents/research/clinicalTrial/2026/data/Criterion list.xlsx'

for dis in disease_lst:
    df = pd.read_excel(input_path, sheet_name=dis)
    df['nct_id'] = df['nct_id'].replace("", None).ffill()
    df['Type']   = df['Type'].replace("", None).ffill()
    output_path = f'/Users/whitneyyu/Documents/research/clinicalTrial/2026/data/{dis}.csv'
    df.to_csv(output_path, index=False)
    print(len(df))
    print(df['nct_id'].nunique())
```
